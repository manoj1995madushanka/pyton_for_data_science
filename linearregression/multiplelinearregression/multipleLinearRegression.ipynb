# It is multiple having at least more than one independence features

## there are two independent feature in this dataset called interest_rate and unemployment_rate

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_index = pd.read_csv('economic_index.csv')
df_index.head()
```

```python
# drop unnecessary columns
df_index.drop(columns=['Unnamed: 0', 'year', 'month'], axis=1, inplace=True)
df_index.head()
```

```python
# check null values
df_index.isnull().sum()
```

```python
# Let's do some visualization
import seaborn as sns

sns.pairplot(df_index)
```

```python
df_index.corr()
```

```python
# visualize datapoints more closely
# negative correlation
plt.scatter(df_index['interest_rate'], df_index['unemployment_rate'], color='r')
plt.xlabel('Interest Rate')
plt.ylabel('Unemployment Rate')
```

```python
# positive correlation
plt.scatter(df_index['interest_rate'], df_index['index_price'], color='r')
plt.xlabel('Interest Rate')
plt.ylabel('Index Price')
```

```python
# Independent and Dependent feature
X = df_index.iloc[:, :-1]  # df_index[['interest_rate'],['unemployment_rate']] do the same thing
y = df_index.iloc[:, -1]  # index_price
```

```python
X.head()
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import seaborn as sns

sns.regplot(x=df_index['interest_rate'], y=df_index['index_price'])
```

```python
sns.regplot(x=df_index['interest_rate'], y=df_index['unemployment_rate'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
# Cross validation model
from sklearn.model_selection import cross_val_score

validation_score = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=3)
```

```python
validation_score
```

```python
np.mean(validation_score)
```

```python
## Prediction

y_pred = regression.predict(X_test)
print(y_pred)
```

```python
# Performance metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
## R square
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
print(score)

# display adjusted R squared
print(1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1))
```
